```python
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn import metrics
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

iris = datasets.load_iris()

x_iris = iris.data 
y_iris = iris.target
```

```python
x = pd.DataFrame(iris.data, columns = ['Sepal Lenght', 'Sepal Width', 'Petal Length', 'Petal Width'])
y = pd.DataFrame(iris.target, columns = ['Target'])
x.head(5)
```

```python
plt.scatter(x['Petal Length'], x['Petal Width'], c = 'green')
plt.xlabel('Petal Length', fontsize = 10)
plt.ylabel('Petal Width', fontsize = 10)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, max_iter=1000, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model = KMeans(n_clusters=3,max_iter=10000)
model.fit(x)
y_labels = model.labels_
```

```python
y_kmeans = model.predict(x)
colormap = np.array(['r','b','purple','black','g'])
plt.scatter(x['Petal Length'], x['Petal Width'],c=colormap[y_labels])
plt.xlabel('Petal Length', fontsize = 10)
plt.ylabel('Petal Width', fontsize = 10)
```

```python
accuracy = metrics.adjusted_rand_score(y_iris,y_kmeans)
print(accuracy)
```
